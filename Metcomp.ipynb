```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np


def generar_muestra(n):
    Lista = []

    for i in range(n):
        u = rnd.random()
        v=rnd.random()

        if 0 <= u <= 0.5:
            x = np.log(2 * v) / 2
        else:
            x = -np.log(2 * (1 - v)) / 2
        Lista.append(x)

    return Lista

```

```python
# Ejemplo de uso
n = 1000000 # NÃºmero de muestras deseadas
G = generar_muestra(n)
```

```python
plt.hist(G,bins=100,color='lightblue')
```

```python
plt.plot(G,color='lightblue')
```

```python
def cdf(muestras):
  n = len(muestras)
  x = np.sort(muestras)
  y = np.arange(0,n)/n
  return x,y
```

```python
x,y = cdf(G)
```

```python
mu=np.mean(G)
```

```python
plt.step(x,y)
plt.axhline(0.5,color="red")
plt.axvline(mu,color="green")
```
