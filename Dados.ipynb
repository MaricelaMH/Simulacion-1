```python
import random as rnd
import matplotlib.pyplot as plt

def volados(n):
    prob = []
    contador_tres = 0
    for i in range(1, n+1):
        resultado = rnd.random()
        if 0.33 < resultado <= 0.49:
            contador_tres += 1
        probabilidad = contador_tres / i
        prob.append(probabilidad)
    return prob

# Número de intentos (cuántos lanzamientos de moneda queremos simular)
n = 1000

# Calcular la probabilidad de obtener un 3 en una serie de lanzamientos
probabilidades = volados(n)

# Crear lista de iteraciones
iteraciones = list(range(1, n+1))

# Graficar las iteraciones vs las probabilidades
plt.plot(iteraciones, probabilidades)
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de obtener un 3 en lanzamientos de moneda')
plt.show()
```

```python
import numpy as np
np.mean(probabilidades)
```
