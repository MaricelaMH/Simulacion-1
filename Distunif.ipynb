```python
def uni(a,b,n):
    Lista = []
    for i in range(n):
        u=rnd.random()
        x=u*(b-a)+a
        Lista.append(x)
    return Lista
```

```python
G=uni(10,15,50)
print(G)
```

```python
mu=np.mean(G)
print(mu)
```

```python
sigma2=np.var(G)
print(sigma2)
```

```python
plt.hist(G,bins=100,color='lightblue')
```

```python
def cdf(G):
  n = len(G)
  x = np.sort(G)
  y = np.arange(0,n)/n
  return x,y
```

```python
x,y = cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5,color="red")
plt.axvline(mu,color="green")
```
