```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la distribución simetrica (normal estandar)
def pi(x):
    return np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)
```

```python
# Inicializamos variables

n = 10000  # Tamaño de la muestra
sigma = 1
x0 = 0

# Inicializamos nuestra cadena que almacenara nuestras muestras
x = np.zeros(n)
x[0] = x0
```

```python
# Aplicamos el método de Metropolis-Hastings
for t in range(1, n):
    # Generamos un candidato
    y = np.random.normal(x[t-1], sigma)

    # Probabilidad de aceptación
    alpha = min(1, pi(y) / pi(x[t-1]))

    # Criterio de aceptación
    u = np.random.uniform()
    if u < alpha:
        x[t] = y
    else:
        x[t] = x[t-1]
```

```python
plt.figure(figsize=(10, 5))

# Histograma de las muestras
plt.hist(x, bins=50, density=True, alpha=0.6, color='pink')

# Grafica de la distribución
x1 = np.linspace(-4, 4, 1000)
plt.plot(x1, pi(x1), linewidth=2, color='r')

plt.title('Metropolis-Hastings Sampling for Normal Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.show()
```
