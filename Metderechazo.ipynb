```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
def muestra(n):
  lista=[]
  for i in range(n):
    y = rnd.random()
    u = rnd.random()
    if u <= (20*y*(1-y)**3)/(135/64):
      x = y
      lista.append(x)
  return lista
```

```python
G = muestra(10000000)
```

```python
plt.hist(G, bins=500, color='lightblue')
```
