```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
def met12unif(n):
    lista = []
    for _ in range(n):
        suma_uniformes = sum(rnd.random() for _ in range(12))
        x = suma_uniformes - 6
        lista.append(x)
    return lista
```

```python
G = met12unif(100000)
```

```python
plt.hist(G, bins=100, color='lightblue')
```

```python
def cdf(G):
  n = len(G)
  x = np.sort(G)
  y = np.arange(0,n)/n
  return x,y
```

```python
x,y = cdf(G)
```

```python
plt.step(x,y)
```
