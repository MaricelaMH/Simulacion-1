FUNCION ACUMULATIVA (INVERSA)

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ej(n):
    Lista = []
    for i in range(n):
        u=rnd.random()
        if 0 <= u < 0.75:
           x = 5/4-(25/16-2*u)**0.5
        else:
          x=4*u-2
        Lista.append(x)
    return Lista
```

```python
G=ej(1000)
#print(G)
```

```python
mu=np.mean(G)
print(mu)
```

```python
sigma2=np.var(G)
print(sigma2)
```

```python
plt.hist(G,bins=100,color='lightblue')
```

```python
plt.plot(G,color='lightblue')
```

```python
def cdf(G):
  n = len(G)
  x = np.sort(G)
  y = np.arange(0,n)/n
  return x,y
```

```python
x,y = cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5,color="red")
plt.axvline(mu,color="green")
```
