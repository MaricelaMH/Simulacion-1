```python
import random as rnd
#Si usamos numpy , ya no es necesario mat
import numpy as np
import matplotlib.pyplot as plt
```

Lo siguiente es el NUCLEO del programa

```python
#Hacemos nuestra función
n = 100
lam = 2
L = -1/lam
for i in range(n):
    # Generamos números aleatorios de (0,1)
    u = rnd.random( )
    print(L*np.log(u))
```

Ahora hacemos la función

```python
Lista = []
def expo(lam,n):
    for i in range(n):
        L=-1/lam
        u=rnd.random()
        x=L*np.log(u)
        Lista.append(x)
    return Lista
```

```python
G=expo(0.5,1000)
print(G)
```

Sacamos la media

```python
mu=np.mean(G)
print(mu)
```

Sacamos la varianza

```python
sigma2=np.var(G)
print(sigma2)
```
