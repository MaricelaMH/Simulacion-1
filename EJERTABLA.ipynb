```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ejertab(n):
  Lista = []
  for i in range(n):
    u=rnd.random()
    if 0 <= u < 0.15:
      x=1
    elif 0.15 <= u  < 0.20:
       x=2
    elif 0.20 <= u < 0.55:
        x=3
    elif 0.55 <= u < 1:
        x=4
    Lista.append(x)

  return Lista
```

```python
G=ejertab(1000)
```

```python
mu=np.mean(G)
print(mu)
```

```python
sigma=np.var(G)
print(sigma)
```

```python
plt.hist(G,color='lightblue')
```

```python
plt.plot(G,color='lightblue')
```

```python
def cdf(G):
  n=len(G)
  x=np.sort(G)
  y=np.arange(0,n)/n
  return x,y
```

```python
x,y=cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5,color="red")
plt.axvline(mu,color="green")
```
