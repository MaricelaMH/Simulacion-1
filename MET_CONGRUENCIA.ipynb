# M É T O D O - D E - C O N G R U E N C I A


---

El método de congruencia lineal es un algoritmo utilizado para generar secuencias de números pseudoaleatorios. La idea básica es generar números a partir de una fórmula matemática que utiliza una operación de congruencia(usualmente módulo) sobre un valor previamente generado, conocido como semilla.


---
* PASOS
1. Seleccionar una semilla inicial X_0. Esta semilla es el primer valor de la secuencia de números pseudoaleatorios y es crucial para generar una secuencia única.
2. Elegir los valores de los parámetros a,c y m , estos parámetros determinan la calidad y características de la secuencia de números generados. Es importante seleccionarlos cuidadosamente para evitar ciclos cortos y sesgos.
3. Utilizamos la fórmula de congruencia lineal X_n+1 = (aX_n + c) mod 'm' para generar cada número pseudoaleatorio en la secuencia. Este proceso implica aplicar la fórmula a la semilla inicial y luego a cada número generado para obtener el siguiente número en la secuencia.
4. Repetimos el paso 3 tantas veces como sea necesario para obtener la cantidad deseada de números pseudoaleatorios.

# EJERCICIO 1

* Determine el ciclo o periodo de vida de los siguientes generadores congruenciales.

         (a) xj+1 = (21xj + 15)mod(31) con x0 = 21
         (b) xj+1 = (13xj + 9)mod(128) con x0 = 7
         (c) xj+1 = (17xj )mod(31) con x0 = 23
         (d) xj+1 = (121 + xj )mod(256) con x0 = 17
         (e) xj+1 = (21xj + 15xj−1)mod(64) con x0 = 21 y xj = 43




---

                (a) xj+1 = (21xj + 15)mod(31) con x0 = 21



```python
print("           M É T O D O   D E  C O N G R U E N C I A               ")
print("------------------------------------------------------------------")


from tabulate import tabulate

# Ingresamos los valores deseados para a,c,m y x_n
iter=500
X_n=21
a=21
c=15
m=31
resultados=[]
imp=set()
cdv=0

#Aplicamos el metodo de congruencia
for i in range(iter):
  val= (a)*(X_n)+c
  mod=(val) % m
  div=X_n /m

  if X_n not in imp:
    resultados.append((i,X_n,val,mod,div))
    imp.add(X_n)
    X_n=mod
    cdv+=1

# Imprimimos los resultados es una tabla
print("\n")
print("-"*11, "T A B L A   D E  R E S U L T A D O S", "-"*11)
print("\n")
headers=["n","X_n","a(X_n)+c","a(X_n)+c(mod 'm')","X_n /m"]
print(tabulate(resultados,headers=headers,tablefmt="grid"))
print("\n")
print("*"*60)
print("* El ciclo de vida es :", cdv)
print("*"*60)
```
